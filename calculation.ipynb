### 3 вариант

Уравнение поперечных колебаний стержня имеет вид:

$$EI\frac{\partial ^4 y(x, t)}{\partial x^4} + P\frac{\partial ^2 y(x, t)}{\partial x^2} + m\frac{\partial ^2 y(x, t)}{\partial t^2} = 0$$

где EI - изгибная жёсткость стержня, m - распределённая масса, P - сжимающая нагрузка.

Один конец стержня закреплён жёстко, а другой шарнирно (l - длина стержня):

$$y(x, t) \bigg|_{x=0} = 0 ; \frac{dy(x, t)}{dx} \bigg|_{x=0} = 0$$
$$y(x, t) \bigg|_{x=l} = 0;  \frac{d ^2 y(x, t)}{dx^2} \bigg|_{x=l} = 0$$

#### 1. Проверка операторов на самосопряжённость:

- A = m, очевидно самосопряженный, так как


$ \int\limits_0^l U_imU_jdx - \int\limits_0^l U_jmU_idx = 0 $

- B = 0 - самосопряженный

- $C = \frac{\partial ^4 y(x, t)}{\partial x^4}EI + P\frac{\partial ^2 y(x, t)}{\partial x^2}$

Получаем сумму двух интегралов:

$ \int\limits_0^l U_iCU_jdx - \int\limits_0^l U_jCU_idx = EI\int\limits_0^l (U_iU_j''''-U_i''''U_j)dx + P\int\limits_0^l (U_iU_j''-U_i''U_j)dx $

Промежуточные вычисления:

---

$ \int\limits_0^l U_iU_j''''dx = U_iU_j'''\bigg|_0^l - \int\limits_0^l U_i'U_j'''dx = (U_iU_j''' - U_i'U_j'')\bigg|_0^l + \int\limits_0^l U_i''U_j''dx $

$ \int\limits_0^l U_iU_j''dx = U_iU_j\bigg|_0^l - \int\limits_0^l U_i'U_j'dx $

---

1. $ (U_iU_j'''-U_i'U_j''+U_i''U_j' - U_i'''U_j)\bigg|_0^l = (0U_j'''-U_i'0+0U_j' - U_i'''0) - (0U_j'''-0U_j''+U_i''0 - U_i'''0) = 0 $

2. $ (U_iU_j'-U_i'U_j)\bigg|_0^l = (0U_j'-U_i'0) - (00-00) = 0 $

Следовательно, оператор C - самосопряжённый

#### 2. При отсутствии сжимающей нагрузки и заданных граничных условиях поставить задачу нахождения собственных значений и собственных форм деформации. Найти три первых собственных значения и соответствующие им формы деформации.

$$ EI\frac{\partial ^4 y(x, t)}{\partial x^4} + m\frac{\partial ^2 y(x, t)}{\partial t^2} = 0; P = 0 $$

$ Пусть \ x = \varphi l \ , \ 0 \leq \varphi \leq l \ , \ t = \tau T \ - \ безразмерные \ переменные $

Уравнение примет вид:

$ \large \frac{EI}{l^4}\frac{\partial ^4 y(\varphi, \tau)}{\partial \varphi^4} + \frac{m}{T^2}\frac{\partial ^2 y(\varphi, \tau)}{\partial \tau^2} = 0 $

$ \normalsize \frac{\partial ^4 y(\varphi, \tau)}{\partial \varphi^4} + \frac{ml^4}{EIT^2}\frac{\partial ^2 y(\varphi, \tau)}{\partial \tau^2} = 0, \ где \ \frac{ml^4}{EIT^2} = 1, \ так \ как \ T = l^2\sqrt{\frac{m}{EI}} $

$ \large \frac{\partial ^4 y(\varphi, \tau)}{\partial \varphi^4} + \frac{\partial ^2 y(\varphi, \tau)}{\partial \tau^2} = 0 $

$ y(\varphi, \tau) = Y(\varphi)T(\tau) $

$ \normalsize Y''''T + YT'' = 0 \ / \ :YT $

Так как $\varphi \ и \ \tau$ - независимые переменные, то равенство возможно только если обе дроби равны константе

$ \large \frac{Y''''}{Y} = -\frac{T''}{T} = \lambda $

$
 \begin{cases}
   Y'''' - \lambda Y = 0, 
   \\
   T'' + \lambda T = 0.
 \end{cases}
$

`Проблема собственных значений и граничные условия:`

$$ Y'''' - \lambda Y = 0 $$
$$ Y\bigg|_{\varphi=0} = 0 ; \ \frac{\partial Y}{\partial \varphi}\bigg|_{\varphi=0} = 0; \ Y\bigg|_{\varphi=1} = 0;  \ \frac{\partial ^2 Y}{\partial \varphi^2} \bigg|_{\varphi=1} = 0$$

$Пусть \ \lambda = \beta^4 \ ; \ Y = e^{ik\varphi}, \ тогда \ характеристическое \ уравнение \ и \ его \ корни \ будут \ иметь \ вид:$

$ k^4 - \beta = 0; \\ k_{1,2} = \pm\beta \ ; \ k_{3,4} = \pm i\beta$

Тогда уравнение примет вид:

$Y = C_1\sin{\beta\varphi} + C_2\cos{\beta\varphi} + C_3\sinh{\beta\varphi} + C_4\cosh{\beta\varphi}$

$Y' =  \beta(C_1\cos{\beta\varphi} - C_2\sin{\beta\varphi} + C_3\cosh{\beta\varphi} + C_4\sinh{\beta\varphi})$

$Y'' = \beta^2(-C_1\sin{\beta\varphi} - C_2\cos{\beta\varphi} + C_3\sinh{\beta\varphi} + C_4\cosh{\beta\varphi}) $

Ищем $C_{1,2,3,4}$ из граничных условий:

$
 \begin{cases}
   C_2 + C_4 = 0, 
   \\
   C_1 + C_3 = 0,
   \\
   -C_3\sin{\beta} - C_4\cos{\beta} + C_3\sinh{\beta} + C_4\cosh{\beta} = 0,
   \\
   C_3\sin{\beta} + C_4\cos{\beta} + C_3\sinh{\beta} + C_4\cosh{\beta} = 0.
 \end{cases}
$

$
 \begin{cases}
   C_2 = -C_4,
   \\
   C_1 = -C_3,
   \\
   C_3\sinh{\beta} + C_4\cosh{\beta} = 0,
   \\
   C_3\sin{\beta} + C_4\cos{\beta} = 0.
 \end{cases}
$

$ \large  \Rightarrow \frac{C_3}{C_4} = \tan{\beta} = \tan{\beta} $

$
\begin{cases}
C_1 = -\tan{\beta} \tilde C = -\tanh{\beta}\tilde C
\\
C_2 = \tilde C
\\
C_3 = \tan{\beta} \tilde C = \tanh{\beta}\tilde C
\\
C_4 = - \tilde C
\end{cases}
$

$ \normalsize Y = \tilde C(- \tan{\beta}\sin{\beta\varphi} + \cos{\beta\varphi} + \tanh{\beta}\sinh{\beta\varphi} - \cosh{\beta\varphi}) = \frac{\tilde C}{\cos{\beta}}(-\sin{\beta}\sin{\beta\varphi} + \cos{\beta}\cos{\beta\varphi}) - \frac{\tilde C}{\cosh{\beta}}(\cosh{\beta}\cosh{\beta\varphi} - \sinh{\beta}\sinh{\beta\varphi}) = \frac{\tilde C}{2\cos{\beta}}(\cos{\beta(\varphi-1)} + \cos{\beta(\varphi+1)} - \cos{\beta(\varphi-1)} + \cos{\beta(\varphi+1)}) - \frac{\tilde C}{2\cosh{\beta}}(\cosh{\beta(\varphi+1)} + \cosh{\beta(\varphi-1)} - \cosh{\beta(\varphi+1)} + \cosh{\beta(\varphi-1)}) = \frac{\tilde C}{\cos{\beta}}\cos{\beta(\varphi+1)} - \frac{\tilde C}{\cosh{\beta}}\cosh{\beta(\varphi-1)}$

`Получаем:`

$ \large Y = \frac{\tilde C}{\cos{\beta}}\cos{\beta(\varphi+1)} - \frac{\tilde C}{\cosh{\beta}}\cosh{\beta(\varphi-1)}$

Проведём нормировку, исходя из условий: $ \int\limits_0^1 Y_н ^ 2d\varphi = 1$

$ \normalsize \int\limits_0^1 Y_н ^ 2d\varphi = \frac{\tilde C^2}{\cos^2{\beta}} \int\limits_0^1 \cos^2{\beta(\varphi+1)} d\varphi + \frac{\tilde C^2}{\cosh^2{\beta}} \int\limits_0^1 \cosh^2{\beta(\varphi-1)} d\varphi - \frac{2\tilde C^2}{\cos{\beta}\cosh{\beta}} \int\limits_0^1 \cosh{\beta(\varphi-1)}\cos{\beta(\varphi+1)} d\varphi$

Получаем сумму трёх интегралов, $ I_{1, 2, 3} $ соответственно:

- $ \normalsize I_1 = \frac{1}{2} \int\limits_0^1 (1 + \cos{2\beta(\varphi+1)}) d\varphi = (\frac{\varphi}{2} + \frac{\sin{2\beta(\varphi+1)}}{4\beta})\big|_0^1 = \frac{1}{2} + \frac{1}{2\beta}\sin{\beta}\cos{3\beta}$

- $ \normalsize I_2 = \frac{1}{2} + \frac{1}{2\beta}\sinh{\beta}\cosh{\beta}$ - Аналогично

- $ \normalsize I_3 = \int\limits_0^1 \cosh{\beta(\varphi-1)}\cos{\beta(\varphi+1)} d\varphi = \frac{1}{\beta} (\cosh{\beta(\varphi-1)\sin{\beta(\varphi+1)}})\big|_0^1 - \int\limits_0^1 (\sinh{\beta(\varphi-1)\sin{\beta(\varphi+1)}}) d\varphi = \frac{1}{\beta} (\sin{2\beta}-\cosh{\beta}\sin{\beta}+\sinh{\beta}\cos{\beta}) - \int\limits_0^1 \cosh{\beta(\varphi-1)}\cos{\beta(\varphi+1)} d\varphi$

$ \large \frac{2I_3}{\cos{\beta}\cosh{\beta}} = \frac{1}{\beta} (\frac{\sin{2\beta}}{\cos{\beta}\cosh{\beta}}-\tan{\beta}+\tanh{\beta}) = \frac{2\sin{\beta}}{\beta\cosh{\beta}} $

`Следовательно` $\tilde C$ `выражается через $\beta$ как:`

$ \large \tilde C = \pm \sqrt{\frac{2}{\frac{1}{\cos^2{\beta}} + \frac{1}{\cosh^2{\beta}} + \frac{1}{\beta}(\frac{\tan{\beta}\cos{3\beta}}{\cos{\beta}} + \tanh{\beta} - \frac{2\sin{\beta}}{\cos{\beta}})}}$

`Тогда` $Y_н$ `будет иметь вид:`

$ \large Y = \pm \sqrt{\frac{2}{\frac{1}{\cos^2{\beta}} + \frac{1}{\cosh^2{\beta}} + \frac{1}{\beta}(\frac{\tan{\beta}\cos{3\beta}}{\cos{\beta}} + \tanh{\beta} - \frac{2\sin{\beta}}{\cos{\beta}})}} (\frac{1}{\cos{\beta}}\cos{\beta(\varphi+1)} - \frac{1}{\cosh{\beta}}\cosh{\beta(\varphi-1)})$

`Полимиальный вид:`

$
\begin{cases}
P_1 = A_1 + B_1\varphi + C_1\varphi^2 + D_1\varphi^3 + E_1\varphi^4
\\
P_2 = A_2 + B_2\varphi + C_2\varphi^2 + D_2\varphi^3 + E_2\varphi^4 + F_2\varphi^5
\\
P_3 = A_3 + B_3\varphi + C_3\varphi^2 + D_3\varphi^3 + E_3\varphi^4 + F_3\varphi^5 + G_3\varphi^6
\end{cases}
$

Граничные условия и условия ортонормированности форм:

$P_i\big|_{\varphi=0,1} = P_i'\big|_{\varphi=0} = P_i''\big|_{\varphi=1} = 0$

```python
import numpy as np
from numpy import sqrt, cos, cosh, sin, tan, tanh, pi
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

def p1(phi):
        C1 = sqrt(5670/19)
        return C1 * (phi**2 - 5/3*phi**3 + 2/3*phi**4)

def krylov(phi, b):
        return - sqrt(2/(1/cos(b)**2 + 1/cosh(b)**2 + (tan(b)*cos(3*b))/(b*cos(b)) + tanh(b)/b - (2*sin(b))/(b*cosh(b)))) * (1/cos(b)*cos(b*(phi+1)) - 1/cosh(b)*cosh(b*(phi-1)))
```

```python
beta = np.linspace(0, 11, 1000)
eps = 0.01
for k in range(4):
        beta[(beta > pi/2 + pi * k - eps) & (beta < pi/2 + pi *k + eps)] = np.nan
plt.plot(beta, tan(beta), label='tan(beta)')
plt.plot(beta, tanh(beta), label='tanh(beta)')
plt.ylim(-1, 3)
plt.legend(loc="best")
```

```python
def getSolution(a, b):
        eps=0.0001
        x = a
        h = 0.0001
        while x < b:
                if abs(tan(x) - tanh(x)) < eps:
                        break
                x += h
        return round(x, 4)
```

```python
beta1 = getSolution(3, 4)
beta2 = getSolution(6.5, 7.5)
beta3 = getSolution(10, 11)

print("Коэффициенты найдены с точностью 10^-4\n")
print("Для первой формы деформции:")
print(f"lamda1 = {beta1**4}; beta1 = {beta1};\n Y1 = {- sqrt(2/(1/cos(beta1)**2 + 1/cosh(beta1)**2 + (tan(beta1)*cos(3*beta1))/(beta1*cos(beta1)) + tanh(beta1)/beta1 - (2*sin(beta1))/(beta1*cosh(beta1))))} * ({1/cos(beta1)}*cos({beta1}*(phi+1)) - {1/cosh(beta1)}*cosh({beta1}*(phi-1)))\n")
print("Для второй формы деформции:")
print(f"lamda2 = {beta2**4}; beta2 = {beta2};\n Y2 = {- sqrt(2/(1/cos(beta2)**2 + 1/cosh(beta2)**2 + (tan(beta2)*cos(3*beta2))/(beta2*cos(beta2)) + tanh(beta2)/beta2 - (2*sin(beta2))/(beta2*cosh(beta2))))} * ({1/cos(beta2)}*cos({beta2}*(phi+1)) - {1/cosh(beta2)}*cosh({beta2}*(phi-1)))\n")
print("Для третьей формы деформции:")
print(f"lamda3 = {beta3**4}; beta3 = {beta3};\n Y3 = {- sqrt(2/(1/cos(beta3)**2 + 1/cosh(beta3)**2 + (tan(beta3)*cos(3*beta3))/(beta3*cos(beta3)) + tanh(beta3)/beta3 - (2*sin(beta3))/(beta3*cosh(beta3))))} * ({1/cos(beta3)}*cos({beta3}*(phi+1)) - {1/cosh(beta3)}*cosh({beta3}*(phi-1)))\n")
```

```python
x = np.linspace(0, 1, 1000)
plt.plot(x, p1(x), label='Полином')
plt.plot(x, krylov(x, beta1), label='Функция Крылова')
plt.legend(loc="best")
```

```python
def standartDeviation(n, f1, f2, beta):
        phi = 0
        deviation = 0
        for _ in range(n):
                deviation += (f1(phi, beta) - f2(phi)) ** 2
                phi += 1 / n

        deviation = sqrt(deviation / n)
        return deviation
```

```python
print(standartDeviation(1000, krylov, p1, beta1))
```
